#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
df_sinasc = sinasc
df_sinasc
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df_sinasc.loc[df_sinasc['APGAR5'] >= 8, 'CAT_APGAR'] = 'normal'
df_sinasc.loc[(df_sinasc['APGAR5'] <= 7) & (df_sinasc['APGAR5'] >= 6), 'CAT_APGAR'] = 'asfixia leve'
df_sinasc.loc[(df_sinasc['APGAR5']) <= 5 & (df_sinasc['APGAR5'] >= 4), 'CAT_APGAR'] = 'asfixia moderada'
df_sinasc.loc[df_sinasc['APGAR5'] <= 3, 'CAT_APGAR'] = 'asfixia grave'
df_sinasc
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
listagestacao = [['Menos de 22 semanas', '22 a 27 semanas', '28 a 31 semanas'],
                 ['32 a 36 semanas', '37 a 41 semanas', '42 semanas e mais']]

plt.rc('figure', figsize = (20, 10))

fig, axes = plt.subplots(2, 3)

for i in range(len(listagestacao)):
    for j in range(len(listagestacao[i])):
        sns.countplot(ax = axes[i, j]
                      , x = 'APGAR5'
                      , hue = 'SEXO'
                      , data = df_sinasc[df_sinasc['GESTACAO'] == listagestacao[i][j]]
                     ).legend(loc = 'upper left').set(title=listagestacao[i][j])
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tab = pd.crosstab(df_sinasc['GESTACAO'], df_sinasc['APGAR5'])

tab
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tab_pctlin = tab.div(tab.sum(axis = 1), axis = 0)

tab_pctlin
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
tab_pctlin.plot.bar()
```

```python
cat = pd.crosstab(df_sinasc['GESTACAO'], df_sinasc['CAT_APGAR'])

cat_pctlin = cat.div(cat.sum(axis = 1), axis = 0)

cat_pctlin
```

```python
cat_pctlin.plot.bar()
```

### Abaixo das 37 semanas de gestação é possível notar um aumento de casos de asfixia leve e severa, pode-se dizer que em partos considerados prematuros o risco para o bebê aumenta.

```python
sns.displot(data = df_sinasc[df_sinasc['PESO'] <= 2200]
            , x = 'PESO'
            , hue = 'CAT_APGAR'
            , hue_order = ['normal', 'asfixia leve', 'asfixia moderada', 'asfixia grave']
            , bins = 30
            , element = 'step'
            , height = 15
            )
```

```python
listaapgar = [['normal', 'asfixia leve'],
                 ['asfixia moderada', 'asfixia grave']]

plt.rc('figure', figsize = (20, 10))

fig, axes = plt.subplots(2, 2)

for i in range(len(listaapgar)):
    for j in range(len(listaapgar[i])):
        sns.histplot(ax = axes[i, j]
                     , x = 'PESO'
                     , bins = 10
                     , data = df_sinasc[df_sinasc['CAT_APGAR'] == listaapgar[i][j]]
                    ).set(title=listaapgar[i][j])

plt.subplots_adjust(wspace = 0.25, hspace = 0.25)
```

```python
listagestacao = [['Menos de 22 semanas', '22 a 27 semanas'], 
                 ['28 a 31 semanas', '32 a 36 semanas'], 
                 ['37 a 41 semanas', '42 semanas e mais']]

plt.rc('figure', figsize = (20, 20))

fig, axes = plt.subplots(3, 2)

for i in range(len(listagestacao)):
    for j in range(len(listagestacao[i])):
        sns.histplot(ax = axes[i, j]
                     , x = 'PESO'
                     , hue = 'CAT_APGAR'
                     , hue_order = ['normal', 'asfixia leve', 'asfixia moderada', 'asfixia grave']
                     , element = 'step'
                     , data = df_sinasc[(df_sinasc['GESTACAO'] == listagestacao[i][j]) 
                                        & (df_sinasc['PESO'] < 3000)]
                    ).set_title(listagestacao[i][j])
```

#### Dentre os nascidos antes de completar 22 semanas de gestação a asfixia grave é diretamente proporcional ao baixo peso. 
#### Já dentre os com mais de 22 semanas e menos de 27 as asfixias leve e grave são observadas dentre os menores pesos.

#### Portanto há uma correlação, porém o nível de APGAR5 não é a única causa de baixo peso nos recén nascidos.

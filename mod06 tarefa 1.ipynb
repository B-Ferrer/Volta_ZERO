#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')

df_sinasc = sinasc

df_sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
df_sinasc[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']].head()
```

```python
fig = plt.figure(figsize = (10 ,10))

ax1 = fig.add_subplot(3,2,1)
ax1 = df_sinasc['APGAR1'].value_counts().sort_index().plot.bar(xlabel = 'APGAR1')
ax2 = fig.add_subplot(3,2,2)
ax2 = df_sinasc['APGAR5'].value_counts().sort_index().plot.bar(xlabel = 'APGAR5')
ax3 = fig.add_subplot(3,2,3)
ax3 = df_sinasc['SEXO'].value_counts().sort_index().plot.bar(xlabel = 'SEXO')
ax3.tick_params(axis="x", rotation = 30)
ax4 = fig.add_subplot(3,2,4)
ax4 = df_sinasc['GESTACAO'].value_counts().sort_index().plot.bar(xlabel = 'GESTACAO')
ax5 = fig.add_subplot(3,2,5)
ax5 = df_sinasc['ESTCIVMAE'].value_counts().sort_index().plot.bar(xlabel = 'ESTCIVMAE')
plt.subplots_adjust(wspace=.2, hspace=.3)

```

 Para testar as duas maneiras de gerar gráficos de barra

```python
plt.close('all')

plt.rc('figure', figsize = (10, 10))

fig, axes = plt.subplots(2, 2)
sns.countplot(ax = axes[0, 0], x = 'APGAR1', data = df_sinasc)
sns.countplot(ax = axes[0, 1], x = 'APGAR5', data = df_sinasc)
sns.countplot(ax = axes[1, 0], x = 'SEXO', data = df_sinasc)
sns.countplot(ax = axes[1, 1], x = 'ESTCIVMAE', data = df_sinasc)
plt.subplots_adjust(wspace=.3, hspace=.2)
```

```python
plt.close('all')

sns.countplot(x = 'GESTACAO', data = df_sinasc).tick_params(axis = 'x', rotation = 30)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')

sns.barplot(x = 'CONSULTAS', y = 'APGAR5', data = df_sinasc)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')

plt.rc('figure', figsize = (10, 10))

fig, axes = plt.subplots(2, 2)
sns.countplot(ax = axes[0, 0], x = 'SEXO', data = df_sinasc)
sns.countplot(ax = axes[0, 1], x = 'GESTACAO', data = df_sinasc).tick_params(axis = 'x', rotation = 30)
sns.countplot(ax = axes[1, 0], x = 'ESTCIVMAE', data = df_sinasc)
sns.countplot(ax = axes[1, 1], x = 'CONSULTAS', data = df_sinasc)
plt.subplots_adjust(wspace=.3, hspace=.5)
```
